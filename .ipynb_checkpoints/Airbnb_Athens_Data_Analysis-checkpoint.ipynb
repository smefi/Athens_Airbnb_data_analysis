```python
import os
os.environ["MODIN_ENGINE"] = "dask" 
from distributed import Client
client= Client(memory_limit='8GB')
import modin.pandas as pd
#import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

#use of modin to get advantage of multi-core processor.
#working on i7 with 16 gb ram
```

```python
pd.options.display.max_columns = None
pd.options.display.max_rows = None
df_listings = pd.read_csv('./airbnb_athens/listings_detailed.csv',low_memory=False, nrows =5000)
df_listings.head()

#use of half of the dataset, memory cannot handle the full dataset, fails.
```

```python
missing_30 = df_listings.columns[df_listings.isnull().mean()  > 0.3] #dropping columns with more than 30% nan values
df_listings.drop(list(missing_30),axis=1, inplace = True) # trexw auto gia na tis diwksw oles. 
```

```python
df_listings.describe()  

#50% of listings have only one bedroom. 
#75% of listings only one bathroom, two bedrooms.

#50% of hosts have 1 or 2 listings (host_total_listings_count)
```

```python
#missing values. dropping columns with more than 30% missing values
df_listings_missing_values_Columns = df_listings.columns[df_listings.isnull().mean()!=0] 
```

```python
df_listings [df_listings_missing_values_Columns]
```

```python
#since we are not going to do any NLP at the moment we will remove columns with free text (do this before handling missing data)
df_listings.drop(['name','summary','space','description','neighborhood_overview','transit'],axis=1, inplace=True)
```

```python

df_listings.drop(['listing_url','scrape_id','picture_url','host_id'], axis=1, inplace = True) #drop unmeaningfull columns like  url
df_listings.drop('last_scraped',axis=1, inplace= True)
df_listings.drop(['host_thumbnail_url','host_picture_url'],axis=1, inplace= True)
```

```python
df_listings_missing_values_Columns = df_listings.columns[df_listings.isnull().mean()!=0] #exw diwksei osa exoyn panw apo 30% null alla prepei na gemisw tis ipoloipes poy exoyn nan values
df_listings [df_listings_missing_values_Columns]
```

```python
#filling missing values. numeric with mean, categorical with mode
```

```python
df_listings.host_location.mode() 
df_listings.host_location = df_listings.host_location.fillna(df_listings.host_location.mode()[0])
df_listings.host_response_time = df_listings.host_response_time.fillna(df_listings.host_response_time.mode()[0])
df_listings.host_neighbourhood = df_listings.host_neighbourhood.fillna(df_listings.host_neighbourhood.mode()[0])
df_listings.neighbourhood = df_listings.neighbourhood.fillna(df_listings.neighbourhood.mode()[0])
df_listings.city = df_listings.city.fillna(df_listings.city.mode()[0])
df_listings.zipcode = df_listings.zipcode.fillna(df_listings.zipcode.mode()[0])
df_listings.market = df_listings.market.fillna(df_listings.market.mode()[0])

df_listings['host_response_rate']= df_listings['host_response_rate'].str.replace("%","")
df_listings['host_response_rate'] = df_listings['host_response_rate'].fillna(df_listings['host_response_rate'].mean())
df_listings['host_acceptance_rate']= df_listings['host_acceptance_rate'].str.replace("%","").astype('float')
df_listings['host_acceptance_rate']= df_listings['host_acceptance_rate'].fillna(df_listings['host_acceptance_rate'].mean())

df_listings.cleaning_fee= df_listings.cleaning_fee.str.replace("$","").astype('float')
df_listings.cleaning_fee.dtype
df_listings.cleaning_fee = pd.to_numeric(df_listings.cleaning_fee, errors = 'coerce')
df_listings.cleaning_fee.dtype
df_listings['cleaning_fee'] = df_listings['cleaning_fee'].fillna(df_listings['cleaning_fee'].mean()) 

df_listings.bathrooms= df_listings.bathrooms.fillna(df_listings.bathrooms.mean())
df_listings.bedrooms= df_listings.bedrooms.fillna(df_listings.bedrooms.mean())

df_listings.review_scores_rating= df_listings.review_scores_rating.fillna(df_listings.review_scores_rating.mean())
df_listings.review_scores_accuracy= df_listings.review_scores_accuracy.fillna(df_listings.review_scores_accuracy.mean())
df_listings.review_scores_cleanliness= df_listings.review_scores_cleanliness.fillna(df_listings.review_scores_cleanliness.mean())

df_listings.review_scores_checkin= df_listings.review_scores_checkin.fillna(df_listings.review_scores_checkin.mean())
df_listings.review_scores_communication= df_listings.review_scores_communication.fillna(df_listings.review_scores_communication.mean())

df_listings.review_scores_location= df_listings.review_scores_location.fillna(df_listings.review_scores_location.mean())
df_listings.review_scores_value= df_listings.review_scores_value.fillna(df_listings.review_scores_value.mean())
df_listings.reviews_per_month= df_listings.reviews_per_month.fillna(df_listings.reviews_per_month.mean())
```

```python
pd.to_numeric(df_listings['host_response_rate'], errors = 'coerce')
df_listings['host_response_rate'] = df_listings['host_response_rate'].fillna(df_listings['host_response_rate'].mean()) 
df_listings['host_acceptance_rate'] = df_listings['host_acceptance_rate'].fillna(df_listings['host_acceptance_rate'].mean()) 

df_listings.isnull().mean() #checking if there are still columns with missing values
```

```python
#type conversion so we can succesfully fill the missing values
pd.to_datetime(df_listings.first_review)
pd.to_datetime(df_listings.last_review)

df_listings.first_review = df_listings.first_review.fillna(method='pad')
df_listings.last_review = df_listings.last_review.fillna(method='pad')
```

```python
df_listings.price
```

```python
#removing symbols
df_listings['price'] = df_listings['price'].str.replace("$","")
df_listings['price'] = df_listings['price'].str.replace(",","")
df_listings['price'] = pd.to_numeric(df_listings['price'], errors = 'coerce')
df_listings.extra_people = df_listings.extra_people.str.replace(",","")

df_listings['security_deposit'] = df_listings['security_deposit'].str.replace("$","")
df_listings['security_deposit'] = df_listings['security_deposit'].str.replace(",","")
```

```python
df_listings.columns[df_listings.isnull().mean()>0]
```

```python
df_listings['beds']= df_listings['beds'].astype(float).fillna(df_listings['beds'].mean())
df_listings['security_deposit'] = df_listings['security_deposit'].astype(float)
df_listings['security_deposit'] = df_listings['security_deposit'].fillna(df_listings['security_deposit'].mean())

df_listings['price'] = df_listings['price'].fillna(df_listings['price'].mean())
df_listings['cleaning_fee'] = df_listings['cleaning_fee'].astype(float).fillna(df_listings['cleaning_fee'].mean())

df_listings['host_response_rate'] = df_listings['host_response_rate'].astype(float)
df_listings['host_response_rate'] = df_listings['host_response_rate'].fillna(df_listings['host_response_rate'].mean())
```

```python
#calendar. do prices change based on the month of visit? what the months is best to travel in order to get lower prices?

df_calendar = pd.read_csv('./airbnb_athens/calendar.csv',low_memory=False, nrows=5000)
```

```python
df_calendar['year'] = pd.DatetimeIndex(df_calendar['date']).year
df_calendar['month'] = pd.DatetimeIndex(df_calendar['date']).month

df_calendar.head()
```

```python
df_calendar.isnull().mean()
```

```python
df_calendar['price'] = df_calendar['price'].str.replace("$","")
df_calendar['price'] = df_calendar['price'].str.replace(",","")
df_calendar['price'] = pd.to_numeric(df_calendar['price'], errors = 'coerce')

df_calendar.groupby('month').price.mean().sort_values(ascending=False) #months with lowest and higher prices

```
